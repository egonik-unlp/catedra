```python
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
import ipywidgets as wg
%matplotlib nbagg
```

```python
x = np.linspace(1,10,100)


fig = plt.figure(figsize = ((10,4)))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
label_comp = 'comun,{}'
ax1.plot(x,1.2*x, label = label_comp.format('distinto'))
ax1.legend(loc = 1)
ax1.set_ylim(1,10)
ax2.set_ylim(1,10)
ax2.plot(x,1.4*x,label = label_comp.format('distinto2'))
ax2.legend(loc = 2)




```
