## Gas Ideal
$$PV = nRT$$            
$$P = nRT\frac{1}{V}$$  
## Gas VdW
$$\big(P + \ \frac{n^2a}{V^2} \big) (V\ - \ nb) =\ nRT$$  $$P = \frac{nRT}{V\ -\ nb} \ - \frac{n^2a}{V^2}$$

```python
V = np.linspace(0.06,3,1000)
fig, ax = plt.subplots(1,figsize=(8,4))
plt.suptitle('P vs V')

def update_plot(real1,n1,T1,real2, n2,T2):
    """
    This function is linked to the sliders and it replots the sine wave when the sliders are modified
    """
    ax.clear()
    R = 0.082
    a = 0.0341 
    b = 0.02370
    if not real1:
        P1 = (n1*T1*R)/V
    else:
        P1 = ((n1*R*T1)/(V- n2*b)) - (n1**2*a/(V**2))
    if not real2:
        P2 = (n2*T2*R)/V
    else:
        P2 = ((n2*R*T2)/(V- n2*b)) - (n2**2*a/(V**2))

    units = 'n = {} moles \nT = {} $K$'
    ax.plot(V,P1, label = units.format(n1,T1))
    ax.plot(V,P2, label = units.format(n2,T2), c = 'r' )
    ax.legend(loc=1)
    ax.set_xlabel('$L$')
    ax.set_ylabel('atm')
    ax.set_xlim((0.05,3))
    plt.show()
real1 = wg.Checkbox(value = False, description = 'Gas vDW?')    
n1 = wg.FloatSlider(min= 0.01,max =10, value= 0.1, description = 'nAzul (m)')
T1 = wg.FloatSlider(min= 273,max =500, value= 298 , description = 'TA (K)')
real2 = wg.Checkbox(value = False, description = 'Gas vDW?') 
n2 = wg.FloatSlider(min= 0.01,max =10, value= 0.1, description = 'nRojo (m)')
T2 = wg.FloatSlider(min= 273,max =500, value= 298 , description = 'TR (K)')


wg.interactive(update_plot,real1 = real1, n1 = n1, T1 = T1,real2=real2, n2 = n2, T2 = T2)

left_box = wg.VBox([real1,n1,T1 ])
right_box = wg.VBox([real2,n2, T2])
wg.HBox([left_box, right_box])
```
