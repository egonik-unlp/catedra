# Gases Ideales

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

%matplotlib nbagg
```

$$PV = nRT$$

$$P = nRT\frac{1}{V}$$


```python
V = np.linspace(0.1,3,1000)
fig, ax = plt.subplots(1,figsize=(10,4))
plt.suptitle('P vs V')

def update_plot(n1,T1,n2,T2):
    """
    This function is linked to the sliders and it replots the sine wave when the sliders are modified
    """
    ax.clear()
    R = 0.082 
    P1 = (n1*T1*R)/V
    P2 = (n2*T2*R)/V
    units = 'n = {} moles \nT = {} $K$'
    ax.plot(V,P1, label = units.format(n1,T1))
    ax.plot(V,P2, label = units.format(n2,T2), c = 'r' )
    ax.legend(loc=1)
    ax.set_xlabel('$L$')
    ax.set_ylabel('atm')
    plt.show()
    
n1 = widgets.FloatSlider(min= 0.01,max =10, value= 0.1, description = 'n1 (moles)')
T1 = widgets.FloatSlider(min= 273,max =500, value= 298 , description = 'T1 (K)')
n2 = widgets.FloatSlider(min= 0.01,max =10, value= 0.1, description = 'n2 (moles)')
T2 = widgets.FloatSlider(min= 273,max =500, value= 298 , description = 'T2 (K)')

widgets.interactive(update_plot,n1 = n1, T1 = T1, n2 = n2, T2 = T2)
```
